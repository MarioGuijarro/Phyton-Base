El teorema de los ceros de Bolzano enuncia que dados dos números $a$ y $b$, reales con $a<b$ y sea una función $f$ definida en $\mathbb{R}$ continua en el intervalo cerrado $[a,b]$. Si el signo de $f(a)$ y $f(b)$ es distinto, entonces como mínimo existe un valor $c\in (a, b)$ para el que $f(c)=0$.

Este teorema sirve como base para distintos métodos de resolución de ecuaciones. Uno de los métodos consiste en:
* Verificar que el signo de la función en $a$ y $b$ es distinto.
* Calcular el punto $c$ como el valor medio del intervalo $[a, b]$. Se evalua $f(c)$ y:
    * Si $f(c)$ es cero hemos encontrado una raíz (solución de la ecuación).
    * En caso contrato se verifica si $f(c)$ tiene signo opuesto con $f(a)$ o con $f(b)$ y en función de eso se tiene que volver a definir el intervalo como $[a,c]$ o $[c, b]$.
* Con este nuevo intervalo se vuelve a iniciar el proceso.
* El proceso continua hasta alcanzar la precisión especificada, calculada como la diferencia entre $b-a$. Es decir, si $|b - a| < precision$.
